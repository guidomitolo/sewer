```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
```

```python
users = pd.read_csv('data/user_info.csv', delimiter=',', header=None)
users.columns = ['user', 'followers','no_tweets','since','last_check']
```

```python
users['days'] = (pd.to_datetime(users.last_check) - pd.to_datetime(users.since)).dt.days
users['day_tweets'] = users.sort_values(by='no_tweets').no_tweets / users.sort_values(by='no_tweets').days
```

```python
users
```

```python
fig = plt.figure()
user = users.sort_values(by='followers').user
followers = users.sort_values(by='followers').followers
plt.bar(user, followers, color=(0.2, 0.4, 0.6, 0.6))
plt.xticks(user, rotation=45, ha='right')
plt.grid(axis='y')
plt.title('Millions of followers')
plt.show()
```

```python
fig = plt.figure()
user = users.sort_values(by='days').user
days = users.sort_values(by='days').days
plt.barh(user, days, color='green')
plt.grid(axis='x')
plt.title('Days in twitter')
plt.show()
```

```python
fig = plt.figure()
user = users.sort_values(by='no_tweets').user
tweets = users.sort_values(by='no_tweets').no_tweets
plt.bar(user, tweets,color=['pink', 'red', 'green', 'blue', 'cyan', 'orange'])
plt.grid(axis='y')
plt.xticks(user, rotation = 45, ha='right')
plt.title('Num. of tweets')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
user = users.sort_values(by='day_tweets').user
tweets = users.sort_values(by='day_tweets').day_tweets
plt.bar(user, tweets,color=['red'])
plt.grid(axis='y')
plt.xticks(user, ha='right', rotation = 45)
plt.title('Tweets per day in Twitter')
plt.show()
```
